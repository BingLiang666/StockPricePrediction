### Assign each tweet a stock price on a specific date which is the closest available date right after the date when the tweet is posted, such as assigning tweets posted on 12/24, 12/25, and 12/26 the stock price of 12/27.

```python
import pandas as pd

tweets_df = pd.read_csv('output.csv')
```

```python
tweets_df
```

```python
stocks_df = pd.read_csv('MSFT.csv')
```

```python
stocks_df
```

```python
# Transform the column "Date" to type "datetime" with format YYYY-MM-DD using pandas and store it as a new column "timestamp" to the end of tweets_df
tweets_df['tweet_date'] = pd.to_datetime(tweets_df['Date']).dt.date
```

```python
tweets_df['tweet_date'] = pd.to_datetime(tweets_df['tweet_date'])
```

```python
tweets_df
```

```python
# Transform the column "Date" to type "datetime" using pandas and store it as a new column "date" to the end of stocks_df
stocks_df['stock_date'] = pd.to_datetime(stocks_df['Date'])
```

```python
stocks_df = stocks_df.drop('Date', axis=1)
```

```python
stocks_df
```

```python
tweets_df = tweets_df.drop('Unnamed: 0', axis=1)
```

```python
tweets_df = tweets_df.reset_index()
```

```python
tweets_df
```

```python
tweets_df.head()
```

```python
# create a column with same value "dummy" for both dataframes to enforce cross join on "tweets" and "stockp_price"
tweets_df['cross_join_key'] = 'dummy'
stocks_df['cross_join_key'] = 'dummy'
```

```python
# Cross join two dataframes using the same dummy key "cross_join_key"
merged_df = tweets_df.merge(stocks_df, on='cross_join_key', how='outer')
```

```python
merged_df.shape
```

```python
merged_df.groupby('index').size().describe()
```

```python
# filter the qualified rows where the stock price is the future stock price for each tweet
# where "stock_date" is date of stock price date while "tweet_date" is date of tweet
merged_df = merged_df[merged_df['stock_date'] > merged_df['tweet_date']]
```

```python
merged_df = merged_df.sort_values(['index', 'stock_date'])
```

```python
# only retain the closest available future stock price
merged_df = merged_df.groupby('index').head(1)
```

```python
merged_df.groupby('tweet_date').head(1).head(20)
```

```python
merged_df.to_csv('merged_tweets_and_stock_prices.csv', index=False)
```
