```python
# For roBERTa to process the text:
# 1. Instead of mentioning any user, we will just use "@user"
# 2. Instead of any hyperlink, we will use "http"

def clean_tweet(tweet):

    tweet_words = []

    for word in tweet.split(" "):
        if word.startswith("@") and len(word) > 1:
            word = "@user"

        elif word.startswith("http"):
            word = "http"
        
        tweet_words.append(word)
        
    tweet_proc = " ".join(tweet_words)
    
    return tweet_proc
```

```python
df['Clean_Tweet'] = df['Tweet'].apply(clean_tweet)
df
```

### 1.2 Load model and tokenizer

```python
# Create a function to tokenize the words without removing the punctuations and stop words
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax
import torch
```

```python
roberta = "cardiffnlp/twitter-roberta-base-sentiment"

model = AutoModelForSequenceClassification.from_pretrained(roberta)

tokenizer = AutoTokenizer.from_pretrained(roberta)

labels = ['Negative', 'Neutral', 'Positive']

def get_sentiment(tweet):
    
    # convert the tweet into numbers
    encoded_tweet = tokenizer(tweet, return_tensors='pt')

    # unpack the encoded_tweet and pass it to the model
    output = model(**encoded_tweet)
    
    # convert the tensor to a numpy array
    res = output[0][0].detach().numpy()
    
    # convert scores into probabilities
    res = softmax(res)
    
    # return the label and score
    negative = res[0];
    neutral = res[1];
    positive = res[2];
    
    return negative, neutral, positive
```

```python
from tqdm.notebook import tqdm
import numpy as np

!jupyter nbextension enable --py widgetsnbextension
```

```python
# Split the dataframe into 3000 sub dataframes

sub_dfs = np.array_split(df, 3000)

for i in tqdm(range(len(sub_dfs))):
    sub_dfs[i][['Negative', 'Neutral', 'Positive']] = sub_dfs[i]['Tweet'].apply(get_sentiment).apply(pd.Series)
```

```python
df = pd.concat(sub_dfs, ignore_index=True)
df
```

```python
# create a function to set the value of the Sentiment column
def set_sentiment(row):
    if row['Negative'] > row['Neutral'] and row['Negative'] > row['Positive']:
        return -1
    elif row['Neutral'] >= row['Negative'] and row['Neutral'] >= row['Positive']:
        return 0
    elif row['Positive'] > row['Negative'] and row['Positive'] > row['Neutral']:
        return 1
    else:
        return None
```

```python
df['Sentiment'] = df.apply(lambda row: set_sentiment(row), axis=1)
df
```
